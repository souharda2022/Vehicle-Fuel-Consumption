We have 81 coloum that's why we are using for loop to check how many NAN are in each column

```python
obj = DT.isnull().sum()
drop_col = []
drop_count = []
have_nan =[]
nan_count = []
for key,value in obj.items():
    print(value,"              ",key)
    if value > 10000:
      drop_col.append(key)
      drop_count.append(value)

    elif 0 < value <10000:
      have_nan.append(key)
      nan_count.append(value)
```

```python
print(drop_col)
print(have_nan)
print(len(drop_col))
DT = DT.drop(drop_col,axis=1)
print("Shape after dropping:", DT.shape)
```

```python
x = drop_col+have_nan
y = drop_count+nan_count
```

```python
import matplotlib.pyplot as plt

# Assuming `classes` is a list containing the class labels or categories from the dataset
# and `class_counts` is a corresponding list containing the frequency/count of each class.

# Example data (replace with your actual data)

# Create bar chart
plt.figure(figsize=(10, 8))
plt.bar(x, y, color='skyblue')
plt.xlabel('features')
plt.ylabel('No of NAN')
plt.title('Feature NAN value representation')
plt.xticks(rotation=45)  # Rotate labels for better readability
plt.tight_layout()
plt.show()
```

Feature selection and dropping.




```python
unwanted_col = ["vehicle_id","city_electricity_consumption","highway_electricity_consumption","combined_electricity_consumption","hours_to_charge_120v","hours_to_charge_240v","hours_to_charge_ac_240v"]

DT=DT.drop(unwanted_col,axis=1)
print("Shape after dropping:", DT.shape)
```

```python
DT.info()
```

```python
# Function to classify scores
def classify_score(score):


    if 0 <score < 12:

        return 'Low'
    elif 12<score < 24:

        return 'Medium'

    elif 24<score < 48:

        return 'High'


# Apply the function to create a new column 'Score_Category'
DT['NEw col'] = DT['annual_consumption_in_barrels_ft1'].apply(classify_score)
```

```python
category_to_numeric = {'Low': 1, 'Medium': 2, 'High': 3}
DT['NEw col Numeric'] = DT['NEw col'].map(category_to_numeric)

# Ensure all relevant columns are numeric
numeric_columns = ['NEw col Numeric', 'city_mpg_ft1', 'unrounded_city_mpg_ft1', 'city_gasoline_consumption_cd',
                    'city_utility_factor', 'highway_mpg_ft1', 'unrounded_highway_mpg_ft1',
                   'highway_gasoline_consumption_cd',  'highway_utility_factor']
DT[numeric_columns] = DT[numeric_columns].apply(pd.to_numeric, errors='coerce')

# Compute the correlation matrix
corr_matrix = DT[numeric_columns].corr()

# Plot the correlation matrix
plt.figure(figsize=(12, 8))
sns.heatmap(corr_matrix, annot=True, fmt=".2f", cmap='coolwarm', cbar=True, linewidths=0.5)
plt.title('Correlation Matrix Including NEw col')
```

```python
correlate_value = ["highway_mpg_ft1","unrounded_highway_mpg_ft1","highway_gasoline_consumption_cd","highway_utility_factor"]

DT=DT.drop(correlate_value,axis=1)
print("Shape after dropping:", DT.shape)
```

```python
filtered_rows1 = DT[DT['NEw col'] == 'Medium']

filtered_rows1.shape

var1= filtered_rows1.sample(n=2682)

var1.shape



filtered_rows2 = DT[DT['NEw col'] == 'Low']
var2= filtered_rows2.sample(n=2682)

filtered_rows2.shape

var2.shape

filtered_rows3 = DT[DT['NEw col'] == 'High']
var3= filtered_rows3.sample(n=2682)

var3.shape
```

```python
new_DT = pd.concat([var1,var2,var3],axis=0)

new_DT.tail(10)
```

Categorical encoding(One hot encoding)


```python
from sklearn.preprocessing import LabelEncoder
categorical_columns = ['make', 'model', 'class', 'drive','transmission', 'fuel_type', 'fuel_type_1', 'my_mpg_data',"NEw col"]

# Encode each categorical column with LabelEncoder
for column in categorical_columns:
    encoder = LabelEncoder()
    new_DT[column] = encoder.fit_transform(new_DT[column])

non_numeric_columns = new_DT.select_dtypes(include=['object']).columns.tolist()
if non_numeric_columns:
    raise ValueError(f"The following columns are not encoded properly: {non_numeric_columns}")
```

```python
print(categorical_columns)
```

**split the dataset**

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test =train_test_split(new_DT,new_DT["NEw col"],test_size=0.3,random_state=42)
print(x_train.shape)
print(x_test.shape)
```

**MIN-MAX** **Scaling**

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
scaler.fit(x_train)

X_train_scaled = scaler.transform(x_train)

X_test_scaled = scaler.transform(x_test)

print(X_train_scaled)
print("-----------------")
print(X_test_scaled)
```

**KNN**

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
neigh = KNeighborsClassifier(n_neighbors=3)
neigh.fit(x_train, y_train)

x_train_prediction_knn = neigh.predict(x_train)
train_data_accuracy = accuracy_score(x_train_prediction_knn, y_train)
print("Train Accuracy:",train_data_accuracy)
```

```python
from sklearn.metrics import classification_report
print(classification_report(x_train_prediction_knn, y_train))
```

```python
neigh.fit(x_test, y_test)
```

```python
x_test_prediction_knn = neigh.predict(x_test)
test_data_accuracy_knn = accuracy_score(x_test_prediction_knn, y_test)
print("Test Accuracy:",test_data_accuracy_knn)
```

```python
print(classification_report(x_test_prediction_knn, y_test))
```

**SVM**

```python
from sklearn import svm
clf = svm.SVC()
clf.fit(x_train, y_train)
```

```python
x_train_prediction_svc = clf.predict(x_train)
train_data_accuracy_svc = accuracy_score(x_train_prediction_svc, y_train)
print("Test Accuracy:",train_data_accuracy_svc)
```

```python
print(classification_report(x_train_prediction_svc, y_train))
```

```python
x_test_prediction_svc = clf.predict(x_test)
test_data_accuracy_svc = accuracy_score(x_test_prediction_svc, y_test)
print("Test Accuracy:",test_data_accuracy_svc)
```

```python
print(classification_report(x_test_prediction_svc, y_test))
```
